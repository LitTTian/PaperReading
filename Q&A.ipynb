## 训练相关

#### Batch Size

- `batch_size`过小和过大对训练的影响？

- `batch_size`增大
    - 能充分利用GPU的显存，提高训练速度，但显存占用高，易 OOM
    - 但泛化能力下降，容易过拟合
    - 梯度稳定性高（梯度方差小）
    - 学习率需按比例增大，调参难
    - 对硬件要求高
- `batch_size`减小
    - 慢(难并行) -> 如果是1的话，等价于stochastic gradient descent
    - 显存占用低
    - 泛化能力强，易跳出局部最优
    - 梯度不稳定（梯度震荡，噪声大）
    - 可以使用默认小的学习率，调参简单
    - 对硬件要求低
