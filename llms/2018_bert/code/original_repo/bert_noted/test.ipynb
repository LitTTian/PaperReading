```python
import tensorflow as tf
from pprint import pprint
```

```python
input_ids = tf.constant([ # [batch_size, seq_length, input_nums]
    [[101, 1], [102, 2], [0, 0], [0, 0]],
    [[101, 3], [102, 4], [103, 5], [0, 0]]
])
flat_input_ids = tf.reshape(input_ids, [-1])
pprint(flat_input_ids)
```

```python
one_hot_input_ids = tf.one_hot(flat_input_ids, depth=1024)
pprint(one_hot_input_ids)
print(one_hot_input_ids.shape)
```
