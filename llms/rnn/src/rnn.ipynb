### Simple Recurrent Networks (Elman, 1990)

![picture 11](assets/IMG_0502-163810318.png)  




- at each time step, hidden layer activations are copied to "context" later
  - e.g. $H1, H2, H3$ are copied to $S1, S2, S3$
- hidden layer receives connections from input and context layers
- the inputs are fed one at a time to the network, it uses the context layer to“remember” whatever information is required for it to produce the correctoutput9

### Back Propagation Through Time

<!-- Image from -->
<!-- [LSTM](https://colah.github.io/posts/2015-08-Understanding-LSTMs/) -->
![picture 12](assets/IMG_0502-164110209.png)  



- we can "unroll" a recurrent architecture into an equivalent feedforwaed architecture, with shared weights
- applying backpropagation to the unrolled architecture is reffered to as "backpropagation through time" (BPTT)
- we can backpropagate just one timestep, or a fixed number of timesteps, or all the way back to beginning of the sequence

- Other  Recurrent Architectures
  - add "shortcut" connections from input to output
  - connections from output back to hidden
